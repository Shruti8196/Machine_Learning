```python
#Reading train and test sets
train = pd.read_csv("../data/aps_failure_training_set.csv",skiprows=20)
test = pd.read_csv("../data/aps_failure_test_set.csv",skiprows=20)
```

```python
#Printing shapes of data
tr_shape = train.shape
te_shape = test.shape

print("train: ",tr_shape)
print("test:", te_shape)
print()

#Printing class distribution
print(train['class'].value_counts())
print()
print(test['class'].value_counts())
```

#### (b) Data Preparation
 This data set has missing values. When the number of columns with missing values is significant, discarding them is not a good idea.

In reality, when we have a model and we want to fill in missing values, we do not have access to training
data, so we only use the statistics of test data to fill in the missing values.

#### i. Research what types of techniques are usually used for dealing with data with missing values. Pick at least one of them and apply it to this data in the next steps.

- Deleting
- Predicting
- Imputing


```python
def cleaning(df:pd.DataFrame):
    '''Returns cleaned dataset'''
    df["class"] = df["class"].replace({"pos":1,"neg":0}).astype(int)
    for col,dt in zip(df.columns[1:],df.dtypes[1:]):
        if dt=="object":
            df[col] = df[col].replace("na",np.nan).astype(float)   
    return df

train = cleaning(train)
test = cleaning(test)
```

```python
#Checking for nulls in train set
n = train.isnull().sum()

print(f"Nulls: {round((n.sum()/(tr_shape[0]*(tr_shape[1]-1)))*100,2)}%")
pd.DataFrame(n).T.iloc[:,1:]
```

```python
#Checking for nulls in test set
n = test.isnull().sum()

print(f"Nulls: {round( (n.sum()/ (te_shape[0]*(te_shape[1]-1)))*100 , 2) }%")
pd.DataFrame(n).T.iloc[:,1:]
```

```python
Xtrain = train.drop("class",axis=1)
ytrain = train["class"]
Xtest  = test.drop("class",axis=1)
ytest = test["class"]
```

```python
imputer = SimpleImputer(strategy="mean")
imputer.fit(Xtrain)
```

```python
#Saving the imputer instance
joblib.dump(imputer,"imputer.save")
```

```python
#Loading imputer instance
imputer = joblib.load("imputer.save") 

Xtrain1 = imputer.transform(Xtrain)
Xtest1 = imputer.transform(Xtest)
```

#### ii. For each of the 170 features, calculate the coefficient of variation CV = s/m , where s is sample standard deviation and m is sample mean

```python
df1 = np.concatenate([Xtrain1,Xtest1],axis=0)
y = np.concatenate([ytrain,ytest],axis=0)[:,np.newaxis]
df1 = np.concatenate([df1,y],axis=1)

df1columns = list(train.columns[1:])
df1columns.append("class")
df1 = pd.DataFrame(df1,columns=df1columns)
```

```python
print("Coefficient of Variation")

cv = {}
for col in df1.columns[:-1]:
    cv[col] = df1[col].std(ddof=1)/df1[col].mean()
cv
```

#### iii. Plot a correlation matrix for your features using pandas or any other tool

```python
# Create correlation matrix
corr = df1.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

plt.figure(figsize=(50,50))
sns.heatmap(corr,mask=mask,cmap='coolwarm')
plt.show()
```

#### (c) Train a random forest to classify the data set. Do NOT compensate for class imbalance in the data set. Calculate the confusion matrix, ROC, AUC, and misclassification for training and test sets and report them (You may use pROC package). Calculate Out of Bag error estimate for your random forset and compare it to the test error.

```python
def assess_model(ytrue:list,ypred:list,probas,data="Train"):
    '''This function assesses the model performance.
    Returns: Missclassification Error Rate, Classification Report, Confusion Matrix'''
    
    #printing the missclassification error rate
    print(f"{data} Error Rate:",1-accuracy_score(ytrue,ypred))
    
    #roc_auc_score
    roc_auc = roc_auc_score(ytrue,probas)
    
    #Classification Report
    print("Classification Report(Performance on Test Set):\n\n",classification_report(ytrue,ypred))
    print("----------------------------------------------------------------------------------------")

    
    #Plotting the confusion matrix
    plt.figure(figsize=(5,3.5))
    sns.heatmap(confusion_matrix(ytrue,ypred),annot=True,cmap="Blues", fmt="d", 
                xticklabels = ['neg','pos'], 
                yticklabels = ['neg','pos'])
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.title("Confusion Matrix",pad=15)
    plt.show()
    
    #Displaying the roc_auc_curve
    fpr, tpr, _ = roc_curve(ytrue,probas)
    
    #create ROC curve
    plt.plot(fpr, tpr, 'r', label = f'AUC = {roc_auc}')
    plt.legend(loc = 'lower right')
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.title("ROC Curve")
    plt.show()
    
```

```python
model = RandomForestClassifier(oob_score = True)
model.fit(Xtrain1,ytrain.values)
```

```python
# save
joblib.dump(model, "random_forest_c.joblib")
```

```python
#Loading the model
model = joblib.load("random_forest_c.joblib")

train_probas = model.predict_proba(Xtrain1)[::,-1]
ypred_train = model.predict(Xtrain1)

test_probas = model.predict_proba(Xtest1)[::,-1]
ypred_test = model.predict(Xtest1)
```

```python
#Out of bag error Rate
print("OOB Error Rate:",1-model.oob_score_)
```

```python
#Assessing Model on training data
assess_model(ytrue = ytrain.values,ypred = ypred_train,probas=train_probas,data="train")
```

```python
#Assessing Model on testing data
assess_model(ytrue = ytest.values,ypred = ypred_test,probas=test_probas,data="Test")
```

```python
ytest.value_counts()
```

#### (d) Research how class imbalance is addressed in random forests. Compensate for class imbalance in your random forest and repeat 1c. Compare the results with those of 1c.

```python
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline

smote = SMOTE(sampling_strategy='minority')
model1 = RandomForestClassifier(oob_score=True)

pipeline = Pipeline(steps=[('SMOTE',smote),('RandomForestClassifier',model1)])
pipeline.fit(Xtrain1,ytrain.values)
```

```python
# save
joblib.dump(pipeline, "random_forest_d.joblib")
```

```python
#Loading the model
model = joblib.load("random_forest_d.joblib")

#Out of bag error Rate
print("OOB Error Rate:",1-pipeline[1].oob_score_)
```

**OOB Error Rate is closer to test error than to train error.**

```python
train_probas = model.predict_proba(Xtrain1)[::,-1]
ypred_train = model.predict(Xtrain1)

test_probas = model.predict_proba(Xtest1)[::,-1]
ypred_test = model.predict(Xtest1)
```

```python
#Assessing Model on training data
assess_model(ytrue = ytrain.values,ypred = ypred_train,probas=train_probas,data="train")
```

```python
#Assessing Model on testing data
assess_model(ytrue = ytest.values,ypred = ypred_test,probas=test_probas,data="Test")
```

#### Using SMOTE, the recall on test improved with a decrease in precision. 

Precision means the percentage of your results which are relevant. On the other hand, recall refers to the percentage of total relevant results correctly classified by your algorithm. 

In this case, if the goal is that the **positives should never go undetected, we can afford to have false positives as long as the true positives are detected. In this case, I would say this model performs better at doing its job than the previous one.**

#### (e) XGBoost and Model Trees
In the case of a univariate tree, only one input dimension is used at a tree split. In a multivariate tree, or model tree, at a decision node all input dimensions can be used and thus it is more general. In univariate classification trees, majority polling is used at each node to determine the split of that node as the decision
rule. In model trees, a (linear) model that relies on all of the variables is used to determine the split of that node (i.e. instead of using X j > s as the decision rule, one has P j β j X j > s . as the decision rule). 

Alternatively, in a regression tree, instead of using average in the region associated with each node, a linear regression model is used to determine the value associated with that node. One of the methods that can be used at each node is Logistic Regression. Because the number of variables is large in this problem, one can use L1 -penalized logistic regression at each node. 

You can use XGBoost to fit the model tree. Determine α (the regularization term) using cross-validation. Train the model for the APS data set without compensation for class imbalance.
