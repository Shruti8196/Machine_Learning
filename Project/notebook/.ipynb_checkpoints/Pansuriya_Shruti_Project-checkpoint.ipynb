```python
# Importing Libraries
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")

from PIL import Image
import glob
import shutil

import tensorflow as tf
from tensorflow.keras import layers,models
from tensorflow.keras import optimizers
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img
from tensorflow.keras.layers import Input,Flatten,BatchNormalization,Dropout,Dense
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, CSVLogger
from tensorflow.keras.models import load_model

from sklearn.metrics import classification_report,confusion_matrix
```

```python
def assess_model(ytrue:list,ypred:list):
    
    '''This function assesses the model performance.'''

    print("----------------------------------------------------------------------------------------")
    #Classification Report
    print("Classification Report:\n\n",classification_report(ytrue,ypred))

    print("----------------------------------------------------------------------------------------")
    #Plotting the confusion matrix
    plt.figure(figsize=(13,10))
    sns.heatmap(confusion_matrix(ytrue,ypred),annot=True,cmap="Blues", fmt="d")
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.title("Confusion Matrix",pad=15)
    plt.show()

def plot_train_val_errors(filepath:str):
    
    '''Takes filepath of history.csv as input and plots train test errors'''

    # result=pd.DataFrame(history.history)
    # result.to_csv("inceptionResnetV2_history.csv")

    #Loading the saved results for inceptionResnetV2
    result = pd.read_csv(filepath)
    result["Train Error"] = 1-result["accuracy"]
    result["Validation Error"] = 1-result["val_accuracy"]

    #Plotting test vs validation accuracy
    plt.figure(figsize=(13,7))
    plt.plot(result[['Train Error','Validation Error']])
    plt.xlabel("Epochs")
    plt.ylabel("Error")
    plt.title("Train vs Validation Error",fontsize=16)
    plt.show()
    
    
def get_ytest_labels(testdata,n):
    
    '''Getting labels from testdata'''
    
    import math
    number_of_examples = len(testdata.filenames)
    number_of_generator_calls = math.ceil(n / (1.0 * 10)) 
    test_labels = []

    for i in range(0,int(number_of_generator_calls)):
        test_labels.extend(np.array(testdata[i][1]))

    ytrue = np.array(test_labels).argmax(axis=1)
    return ytrue
```

### (c) Transfer Learning 

i. When dealing with classification of relatively small image datasets, deep net- works may not perform very well because of not having enough data to train them. In such cases, one usually uses transfer learning , which uses deep learning models that are trained on very large datasets such as ImageNet as feature extractors. The idea is that such deep networks have learned to extract meaningful features from an image using their layers, and those fea- tures can be used in learning other tasks. In order to do that, usually the last layer or the last few layers of the pre-trained network are removed, and the response of the layer before the removed layers to the images in the new dataset is used as a feature vector to train one more multiple replacement lay- ers. The dataset in this task has only around 50-60 images per class. Given that we have 20 classes, training a deep network with such a small dataset may not yield desirable results. In this project, you will use pre-trained mod- els EfficientNetB0 and VGG16 . For both pre-trained networks, you will only train the last fully connected layer, and will freeze all layers before them (i.e. we do not change their parameters during training) and use the outputs of the penultimate layer in the original pre-trained model as the features extracted from each image. 

ii. To perform empirical regularization randomly zoo, rotate, flip, and translate images in your training set for image augmentation. You can use various tools to do this, including OpenCV. 


iii. Use ReLU activation functions in the last layer and a softmax layer, along with batch normalization 4 and a dropout rate of 20% as well as ADAM optimizer. Use multinomial cross entropy loss. You can try any batch size, but a batch size of 5 seems reasonable. 

iv. Train the networks ( EfficientNetB0 and VGG16 ) for at least 50 epochs (preferably 100 epochs) and perform early stopping using the validation set. Keep the network parameters that have the lowest validation error. Plot the training and validation errors vs. epochs. 

v. Report Precision, Recall, and F1 score for your model. Remember that this is a multi-class classification problem. 

### Training VGG16

```python
data1 = ImageDataGenerator(
#                           rescale=1./255, 
                          shear_range=0.2,
                          zoom_range=0.2, 
                          rotation_range=0.2,
                          width_shift_range = 0.2,
                          height_shift_range = 0.2,
                          horizontal_flip=True,
                          preprocessing_function=tf.keras.applications.vgg16.preprocess_input
                          )

batch_size = 10
traindata1 = data1.flow_from_directory(directory="../data/train/",
                                     target_size = (224,224),
                                     batch_size = batch_size,
                                     class_mode = 'categorical',
                                     shuffle = True
                                     )
validationdata1 = data1.flow_from_directory(directory="../data/validation/",
                                     target_size = (224,224),
                                     batch_size = batch_size,
                                     class_mode = 'categorical',
                                     shuffle = True
                                     )

testdata1 = data1.flow_from_directory(directory = "../data/test/",
                                   target_size = (224,224),
                                   batch_size = batch_size,
                                   class_mode = 'categorical',
                                   shuffle = True
                                   )
```

```python
from tensorflow.keras.layers import Input,Flatten,BatchNormalization,Dropout,Dense,Conv2D,MaxPooling2D,ReLU
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, CSVLogger

vgg = tf.keras.applications.vgg16.VGG16(include_top=False,input_shape=(224,224,3))
vgg.trainable = False
```

```python
# Adding some custom layers
x = Flatten()(vgg.output)
x = Dense(4096,activation="relu")(x)
x = BatchNormalization()(x)
x = Dropout(0.2)(x)
x = Dense(traindata1.num_classes, activation='softmax')(x)
model1 = tf.keras.Model(inputs=vgg.input, outputs = x)
```

```python
model1.summary()
```

```python
#checkpoints to check performance during training and save model weights/data.
earlystopping = EarlyStopping(monitor='val_accuracy', mode='max', verbose=1, patience=20)
checkpointer = ModelCheckpoint(filepath="vgg_classifier_weights.h5", verbose=1, monitor='val_accuracy', save_best_only=True, save_weights_only=False)
csv_logger = CSVLogger("vgg_classifier.csv", append=True, separator=',')
```

```python
#compiling the model. Optimizer: Adam (lr = 0.0001), Loss: categorical cross entropy, Metrics: accuracy.
opt = optimizers.Adam(learning_rate=0.0001)
model1.compile(optimizer=opt,
              loss='categorical_crossentropy', metrics=['accuracy'])

stepsTrain = (traindata1.n // batch_size) 
stepsVal = (validationdata1.n // batch_size) 
```

```python
#model training for 100 epochs
history1 = model1.fit(traindata1, epochs=100, callbacks=[earlystopping, checkpointer, csv_logger], validation_data=validationdata1,
          steps_per_epoch=stepsTrain, validation_steps=stepsVal)

model1.save("./vgg")
```

```python
# #Saving results
# result=pd.DataFrame(history1.history)
# result.to_csv("vgg_history.csv")
```

```python
#Plotting train vs validation errors
plot_train_val_errors("vgg_history.csv")
```

```python
#Loading the model
model_l = load_model("vgg_classifier_weights.h5")

#Gets loss and accuracy for us
model_l.evaluate(testdata1)
```

```python
#Getting labels from testdata
ytrue = get_ytest_labels(testdata1,n=171)
```

```python
#making predictions on testdata
y_pred = model_l.predict(testdata1)
ypred = y_pred.argmax(axis=1)
```

```python
#Assesing the models performance
assess_model(ytrue,ypred)
```

### EfficientNetB0

```python
data1 = ImageDataGenerator(
                          shear_range=0.2,
                          zoom_range=0.2, 
                          rotation_range=0.2,
                          width_shift_range = 0.2,
                          height_shift_range = 0.2,
                          horizontal_flip=True,
                          preprocessing_function=tf.keras.applications.efficientnet.preprocess_input
                          )

batch_size = 10
traindata2 = data1.flow_from_directory(directory="../data/train/",
                                     target_size = (224,224),
                                     batch_size = batch_size,
                                     class_mode = 'categorical',
                                     shuffle = True
                                     )
validationdata2 = data1.flow_from_directory(directory="../data/validation/",
                                     target_size = (224,224),
                                     batch_size = batch_size,
                                     class_mode = 'categorical',
                                     shuffle = True
                                     )

testdata2 = data1.flow_from_directory(directory = "../data/test/",
                                   target_size = (224,224),
                                   batch_size = batch_size,
                                   class_mode = 'categorical',
                                   shuffle = True
                                   )
```

```python
#Getting the model
ebo = tf.keras.applications.EfficientNetB0(include_top=False,input_shape=(224,224,3),classes=traindata2.num_classes)
ebo.trainable = False
```

```python
# Extending the model
x = Flatten()(ebo.output)
x = Dense(4096,activation="relu")(x)
x = BatchNormalization()(x)
x = Dense(4096,activation="relu")(x)
x = Dropout(0.2)(x)
x = Dense(traindata2.num_classes, activation='softmax')(x)
model2 = tf.keras.Model(inputs=ebo.input, outputs = x)
```

```python
model2.summary()
```

```python
#checkpoints to check performance during training and save model weights/data.
earlystopping = EarlyStopping(monitor='val_accuracy', mode='max', verbose=1, patience=20)
checkpointer = ModelCheckpoint(filepath="EfficientNetB0_classifier_weights.h5", verbose=1, monitor='val_accuracy', save_best_only=True, save_weights_only=False)
csv_logger = CSVLogger("EfficientNetB0.csv", append=True, separator=',')
```

```python
#compiling the model. Optimizer: Adam (lr = 0.0001), Loss: categorical cross entropy, Metrics: accuracy.
opt = optimizers.Adam(learning_rate=0.0001)
model2.compile(optimizer=opt,
              loss=tf.keras.losses.CategoricalCrossentropy(), metrics=['accuracy'])

#Steps per batch
stepsTrain = (traindata2.n // batch_size) 
stepsVal = (validationdata2.n // batch_size) 
```

```python
#model training for 100 epochs
history2 = model2.fit(traindata2, epochs=100, callbacks=[earlystopping, checkpointer, csv_logger], validation_data=validationdata2,
          steps_per_epoch=stepsTrain, validation_steps=stepsVal)

# model1.save("./ebo")
```
